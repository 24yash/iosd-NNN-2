```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from xgboost import XGBRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
# categorical and numerical columns seperation although we got only num cols here
categorical_cols = [cname for cname in X_train.columns if
                    X_train[cname].nunique() < 20 and 
                    X_train[cname].dtype == "object"]
numerical_cols = [cname for cname in X_train.columns if 
                X_train[cname].dtype in ['int64', 'float64']]
my_cols = categorical_cols + numerical_cols
```

```python
my_cols
```

```python
#preprocessing  no need here bz no null but still
numerical_transformer = SimpleImputer(strategy='constant')

# no categorical for us here btw but i like to put it here in pipeline :)
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
])
```

```python
# model = XGBRegressor(n_estimators=120, max_depth=8, random_state=0, learning_rate=0.15) 32.63 mae
# model = XGBRegressor(n_estimators=100, max_depth=8, random_state=0, learning_rate=0.15)
# 32.42
model = XGBRegressor(n_estimators=150, max_depth=8, random_state=0, learning_rate=0.15)
# 32.87 

# Bundle preprocessing and modeling code in a pipeline
clf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_valid)

mae = mean_absolute_error(y_valid, y_pred)
print(mae)
```

```python
from sklearn.metrics import r2_score


r2 = r2_score(y_valid, y_pred)
print("R2 Score: ", r2)
```
